```python
import os
import csv
import pandas as pd
import glob

folder_path = r"C:\Users\Abhigyan Borah\Desktop\dataset transfomers\keystroke_ano"
output_folder = os.path.join(folder_path, "csv_output")
os.makedirs(output_folder, exist_ok=True)

columns = ["Timestamp", "EventType", "Key", "Username"]

# Match ALL files, regardless of extension
txt_files = [f for f in glob.glob(os.path.join(folder_path, "*")) if os.path.isfile(f)]

if not txt_files:
    print("⚠️ No files found in folder.")
else:
    for txt_file in txt_files:
        print(f"Processing: {os.path.basename(txt_file)}")
        cleaned_data = []

        with open(txt_file, 'r', encoding='utf-8') as f:
            reader = csv.reader(f)
            for row in reader:
                if len(row) == 2:
                    cleaned_data.append([row[0], row[1], None, None])
                elif len(row) == 4:
                    cleaned_data.append(row)
                else:
                    print(f"⚠️ Skipping malformed row: {row}")

        if cleaned_data:
            df = pd.DataFrame(cleaned_data, columns=columns)
            base_name = os.path.splitext(os.path.basename(txt_file))[0]
            csv_path = os.path.join(output_folder, base_name + ".csv")
            df.to_csv(csv_path, index=False)
            print(f"✅ Saved: {csv_path}")
        else:
            print(f"⚠️ No valid data in: {txt_file}")
```
